```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv("/content/frameTime Single.csv", sep=';')
```

```python
print(df.columns)
```

```python
df['Frame Offset'] = df['Frame Offset'].astype('float64')
df[' Frame Index'] = df[' Frame Index'].astype('float64')
df[' Frame Time (ms)'] = df[' Frame Time (ms)'].astype('float64')
df[' Time from first frame (ms)'] = df[' Time from first frame (ms)'].astype('float64')
```

```python

max_frame_time = df[' Frame Time (ms)'].max()

min_time_from_first_frame = df[' Time from first frame (ms)'].min()

average_frame_index = df[' Frame Index'].mean()

median_frame_offset = df['Frame Offset'].median()

tiempo_transcurrido = df[' Time from first frame (ms)']
tiempo_ejecucion = df[' Frame Time (ms)']

plt.figure(figsize=(10, 6))
plt.plot(tiempo_transcurrido, tiempo_ejecucion, marker='o', linestyle='-', markersize=2)
plt.title('Execution time of each frame')
plt.xlabel('Time from first frame (ms)')
plt.ylabel('Execution time (ms)')
plt.grid(True)
plt.show()

print("The maximum frame execution time is:", max_frame_time, "ms")
print("The minimum time from the first frame is:", min_time_from_first_frame, "ms")
print("The average frame index is:", average_frame_index)
print("The median frame offset is:", median_frame_offset)
```
